```python
# Checking missing data and values
print(df.isnull().sum())
```

```python
# filling the missing data set
df['peak-rpm']=df['peak-rpm'].fillna(df['peak-rpm'].mean(),)
df['horsepower']=df['horsepower'].fillna(df['horsepower'].mean(),)
df['stroke']=df['stroke'].fillna(df['stroke'].mean(),)
df['bore']=df['bore'].fillna(df['bore'].mean(),)
df['num-of-doors']=df['num-of-doors'].fillna(df['num-of-doors'].mean(),)
df['normalized-losses']=df['normalized-losses'].fillna(df['normalized-losses'].mean(),)
```

```python
# filling the missing data set
df['price']=df['price'].fillna(df['price'].mean())
```

```python
# check if the change has been made 
print(df.isnull().sum())
```

```python
# Histogram of engine size
plt.hist(df['engine-size'],bins=30,color='skyblue',edgecolor='black')
plt.title('The distribution of engine size', fontsize=16,fontweight='bold')
plt.xlabel('engine-size')
plt.ylabel('frequency')
plt.show()
```

```python
# Histogram of body style
plt.hist(df['body-style'],bins=30,color='skyblue',edgecolor='black')
plt.title('The distribution of engine size', fontsize=16,fontweight='bold')
plt.xlabel('Body style')
plt.ylabel('Frenquency')
plt.show()
```

```python
# scatter plot: engine size vs.price
plt.scatter(df['engine-size'],df['price'],color='skyblue',edgecolor='black')
plt.title('The distribution of engine size', fontsize=16,fontweight='bold')
plt.xlabel('engine size')
plt.ylabel('price')
plt.show()
```
